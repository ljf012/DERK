```python
import pandas as pd
import gzip
import json



df = pd.read_csv('./ml-1m/movies.dat', sep='::', names=['movieId', 'title', 'genres'])
df
```

```python
df['genres'] = df['genres'].str.split('|')
df = df.explode("genres")
df
```

```python
df['genres'].value_counts()
```

```python
ca_a = df['genres'].value_counts()
ca = list(ca_a.index)

num = list(ca_a.values)

id = list(range(len(ca)))
remap = {
    "category" : ca,
    "cate_count" : num,
    "cate_id" : id
}
ca_remap = pd.DataFrame(remap)

# ca_remap = ca_remap[~ca_remap['cate_count'].isin([1])] # 取反，删除只出现过1次的类别

ca_remap.to_csv('./category_list.txt', sep='\t', index=False)

ca_remap
```
